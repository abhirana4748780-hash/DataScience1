```python
#Write a Python program to calculate the factorial of a number

def factorial(n):
    if n < 0:
        return "Number is not get negative"
    elif n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)

num = int(input("Enter the Number"))
print(f"the Factorial number {num} is: {factorial(num)}")    
```

```python
# Write a Python program to check if a year is a leap year 

year = int(input("enter the number: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"Year is leap year: {year}")
else:
    print(f"Year is not leap year: {year} ")
```

```python
# Write a Python program to reverse a string

string = input("Enter the string: ")
rev_string = string[::-1]
print(f"the string is {string}")
print(f"the reversed string is {rev_string}")
```

```python
# Write a Python program to check if a string is a palindrome
string = input("Enter the string: ")
rev_string = string[::-1]

if string == rev_string:
    print("this is palindrome")
else:
    print("this is not the palindorme")
```
