```python
# Q4 Define a function that takes three named arguments: first_name, last_name, and age, and returns a formatted string. Test the function with different inputs.
def names(first_name,last_name,age):
    return f"My name is {first_name} {last_name} and my age is {age}"

n1 = names(first_name="Abhinav",last_name="Chauhan",age=22)
n2 = names(first_name="Bhanu",last_name="Partap",age=21)

print(n1)
print(n2)
```

```python
# Q5 Define a function that takes a variable number of integer arguments and returns their product. Test the function with different inputs
def func(*args):
    product = 1
    for num in args:
        product *= num
    return product  

f1 = func(2,3,4,5,6,7,8)
print(f1)
```

```python
# Q6 Define a function that contains another function inside it. The outer function should take two integers as input and return the result of the inner function, which multiplies the two integers. Test the function with different inputs.

def outer(x,y):
    def inner(a,b):
        return a * b
    return inner(x,y)

print(outer(4,5))
print(outer(8,9))
```

```python
# Q7 Define a function that takes a single integer as input and returns the integer squared, cubed, and raised to the power of four. Test the function with different inputs.

def power(x):
    return x **2 , x**3,x**4

print(power(2))
print(power(6))
```

```python
# Q8 Define a recursive function that calculates the factorial of a given number. Test the function with different inputs.
def fact(n):
    if n == 1:
        return 1
    else:
        return n * fact(n-1)

print(fact(5))
print(fact(6))
```

```python
# Q9 Define a lambda function that takes two integers as input and returns their sum. Test the lambda function with different inputs.
add_num = lambda x,y : x + y

print(add_num(5,6))
```

```python
# Q10 Use the map function to apply a lambda function that squares each number in a list of integers. Test with different lists
number = [1,2,3,4,5,6,7,8,9]

num = list(map(lambda x : x**2 ,number))

print(num)
```

```python
# Q11 Use the filter function to filter out all odd numbers from a list of integers. Test with different lists.
number = [1,2,3,4,5,6,7,8]
num = list(filter(lambda x : x%2 ==0 ,number))
print(num)
```

```python
# Q12 Define a decorator function that prints 'Executing function...' before executing a function and 'Function executed.' after executing it. Apply this decorator to a function that takes a list of integers and returns their sum. Test the decorated function with different lists.

def my_decorator(func):
    def wrappper(*args,**krgs):
        print("Function executed")
        result = func(*args,**krgs)
        print("Function dexecuted")
        return result
    return wrappper

@my_decorator
def sum_lst(lst):
    return sum(lst)

print(sum_lst([1,2,3,4,5,6]))
```

```python
# Q13 Define a function that takes variable-length arguments and keyword arguments and prints them. Test the function with different inputs.
def func(*args,**kwrgs):
    print("args", args)
    print("kwrgs",kwrgs)

print(func(1,2,3,a='apple',b='ball'))
print(func("hello","world",a=20,b=30))
```

```python
# Q14 Define a higher-order function that takes a function and a list of integers as arguments, and applies the function to each integer in the list. Test with different functions and lists.

def func1(fun,lst):
    return [fun(x) for x in lst]

print(func1(lambda x: x ** 2, [1, 2, 3, 4, 5]))
print(func1(lambda x: x + 2, [1, 2, 3, 4, 5]))
```
