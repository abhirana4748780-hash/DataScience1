```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_end = [10, 12, 14, 15, 16, 18, 1000]
df = pd.DataFrame({"Values":data_end})
```

```python
data_percentile = df["Values"].quantile([0.25,0.5,0.75])
print(data_percentile)
```

```python
sns.histplot('data_percentile')
```

```python
plt.bar('data_percentile',10)
```

```python
data_end = np.percentile(df,[25,50,75])
print(data_end)
```
